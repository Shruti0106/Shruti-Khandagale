```python
import numpy as np
import pandas as pd
```

```python
# Import dataset
ds = pd.read_csv("data_1.csv")
ds
```

```python
# handling missing value using meann and scikit learn


from sklearn.impute import SimpleImputer
import numpy as np

si = SimpleImputer(missing_values=np.nan, strategy='mean')
ds[["Age", "Salary"]] = si.fit_transform(ds[["Age", "Salary"]])
```

```python
ds
```

```python
# encoding categical column (one hot encoder)

from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False) 
encoded_data = ohe.fit_transform(ds[['Country']])
print(encoded_data)
```

```python
# make column headers and make a dataframe for array

column_names = ohe.get_feature_names_out(['Country'])
encoded_df = pd.DataFrame(encoded_data, columns = column_names)
encoded_df
```

```python
# Joining original and new dataset

ds = pd.concat([ds,encoded_df], axis = 1)
ds
```

```python
ds = ds.loc[:, ~ds.columns.duplicated()]
ds
```

```python
# droping categarical column

ds = ds.drop(['Country'], axis = 1)
ds
```

```python
# Label encoding

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
ds['Purchased'] = le.fit_transform(ds['Purchased'])
```

```python
ds
```

```python
# Feature scaling(Reduce laarger data values to its smller data values)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
ds[['Age','Salary']] = sc.fit_transform(ds[['Age','Salary']])
```

```python
ds
```

```python
# Divide dataset as independent (input) and dependaent (output) variables

x = ds[['Age','Salary','Country_France','Country_Germany','Country_Spain']]
y = ds[['Purchased']]
```
