```python
import pandas as pd
```

```python
# Sample data
data = {
     'CustomerID':['C001','C002','C003','C004','C005','C006','C007','C008'],
     'Education_Level':['High school','Bacholers','Masters','PhD','PhD','Bacholers','High school','Masters'],
     'Product_Quality':['Low','Medium','High','Medium','Low','High','Medium','High'],
     'Satisfaction_Level':['Poor','Average','Excellent','Good','Average','Excellent','Poor','Good']
 }

df = pd.DataFrame(data)
```

```python
df
```

```python
#Define the order for each ordinal features 

education_order = ['High school', 'Bacholers','Masters', 'PhD']
product_qaulity_order = ['Low','Medium','High']
satisfaction_order = ['Poor','Average','Good','Excellent']
```

```python
education_order
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
encoder = OrdinalEncoder(categories=[education_order,
                                    product_qaulity_order,satisfaction_order])
```

```python
encoder
```

```python
#Apply encoding (skip customer id it is identifier)

df[['Education_Level','Product_Quality','Satisfaction_Level']] = encoder.fit_transform(df[['Education_Level','Product_Quality','Satisfaction_Level']])
```
