```python
# import libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Line plot
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [12,34,43,76,50]
plt.plot(x,y)
plt.xlabel("data1")
plt.ylabel("data2")
plt.show()
```

```python
# 2. Bar plot
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.DataFrame ({
    "class" : ['A','B','C','D','E'],
    "students" : [45,57,45,23,67]
})
sns.barplot(x = 'class', y = 'students',data = data)
plt.title("Barplot")
plt.xlabel("class")
plt.ylabel("students")
plt.show()
```

```python
class A having more students followed by class D and E
```

```python
# 3. Count plot
sns.countplot(x = 'class', data = data)
plt.title("Countplot")
plt.show()
```

```python
     #  --------------------UNIVARIATE ---------------------
# 4. Histogram

import numpy as np
data = np.random.randn(100)
plt.hist(data,bins = 10)
plt.title("Histogram")
plt.show()
```

```python
# 5. KDE plot
sns.kdeplot(data,shade = True)
plt.title("KDE Plot")
plt.show()
```

```python
# or using seaborn
sns.kdeplot(data, bins = 10, kde = True)
plt.title("Histigram with kde")
plt.show()
```

```python
#sample data
data = [10,20,30,35,46,57,53,57]
# create kde plot 
sns.kdeplot(data)
#add title and labels
plt.title("kde plot")
plt.xlabel("value")
plt.ylabel("density")
# show the plot
plt.show()
```

```python
# 6. box plot
sns.boxplot(y = data)
plt.title('Box Plot')
plt.show()
```

```python
# 7. violin plot (Boxplot kde)
sns.violinplot(y = data)
plt.title("violin plot")
plt.show()
```

```python
# 8. Scatter plot (to show connection, correlation between two variables)

x = [1,2,3,4,5],
y = [1,2,3,4,5]
plt.scatter(x,y)
plt.title("scatter plot")
plt.show()
```

```python
# seaborn way library
sns.scatterplot(x = x, y = y)
plt.title("scatter plot")
plt.show()
```
