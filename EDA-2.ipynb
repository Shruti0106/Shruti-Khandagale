```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd.read_csv("employee_dataset_for_eda.csv")
ds
```

```python
ds.info()
```

```python
# Employee count per department

sns.countplot(data = ds, x = 'Department', palette = 'Set2')
plt.title("Employee couont per department")
```

Fianance department having more employee count followed by marketing and engineering department

```python
# Gender Distribution

sns.countplot(data = ds, x = "Gender", palette = "coolwarm")
plt.title("Gender Distribution")
```

Female employees are more than male employees

```python
# Employee location distribution 

sns.countplot(data = ds, x = "Location", palette = "Spectral")
plt.title("Number of Employees by distribution")
plt.xticks(rotation=45)
```

Seattle has more employees than New york and San Fransisco

```python
# Age distribution

sns.histplot(ds['Age'], bins=10, kde=True)
plt.title("Age distribution")
```

Employees of age between 45 to 50 are more followed by age of 25 to 30 employees

```python
# Year of experiance

sns.histplot(ds['Years_of_Experience'], bins = 15, kde = True)
plt.title('Years of Experience')
```

```python
# Monthly salary
sns.boxplot(data = ds, y='Monthly_Salary', color='lightblue')
plt.title('Monthly Salary')
```

Minimum salary is 20,000 
Maximum salary is 1,20,000
Middle/Avg salary is 75,000

```python
# Average salary by department
sns.barplot(data=ds, x='Department', y='Monthly_Salary', estimator=np.mean, palette='viridis')
plt.title('Average salary by department')
```

We can analysis the avg salary from each colored block's black line. The avg maximum salary is around 69,000 of Fianance dep and avg minimum salary is 60,000 of Sales dep.

```python
# Performance score by department
sns.boxplot(data = ds, x='Department', y = 'Performance_Score', palette='pastel')
plt.title('Performance score by department')
```

```python
Marketing department has more performace score than
```

```python
# Correlation Matrix
sns.heatmap(ds[['Age','Years_of_Experience','Monthly_Salary','Performance_Score']].corr(),annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
```

```python
# Salary vs. Experience based on Department scatter plot
sns.scatterplot(data=ds, x='Years_of_Experience', y='Monthly_Salary', hue='Department')
plt.title('Salary vs. Experience based on Department')
```

```python
# Salary Distribution by gender
sns.violinplot(data=ds, x='Gender', y='Monthly_Salary',palette='muted')
plt.title('Salary Distribution by gender')
```

Males have more monthly salary than females
